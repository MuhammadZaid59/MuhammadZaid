You work for a production company that specializes in nostalgic styles. You want to do some research on movies released in the 1990's. You'll delve into Netflix data and perform exploratory data analysis to better understand this awesome movie decade!

You have been supplied with the dataset `netflix_data.csv`, along with the following table detailing the column names and descriptions. Feel free to experiment further after submitting!

## The data
### **netflix_data.csv**
| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |
| `type` | Type of show |
| `title` | Title of the show |
| `director` | Director of the show |
| `cast` | Cast of the show |
| `country` | Country of origin |
| `date_added` | Date added to Netflix |
| `release_year` | Year of Netflix release |
| `duration` | Duration of the show in minutes |
| `description` | Description of the show |
| `genre` | Show genre |

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the Netflix CSV file into a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

# Selecting relevant columns: 'title', 'release_year', 'duration', and 'genre'
Sorted_df = netflix_df[['title', 'release_year', 'duration', 'genre', 'type']]
print(Sorted_df)

# Filtering movies released between 1990 and 1999
movies = Sorted_df[(Sorted_df['type'] == 'Movie') & (Sorted_df['release_year'] >= 1990) & (Sorted_df['release_year'] < 2000)]
print(movies)

# Extracting 'duration' column for the histogram
data = movies['duration']

# Create histogram for movie durations
plt.hist(data, bins=30, color='blue', edgecolor='black')

# Adjust the X-axis ticks to show intervals of 20 minutes (from 0 to 200 minutes)
plt.xticks(range(0, 201, 20))

# Add labels and title to the plot
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.title('Frequency of Movie Durations (1990-1999)')

# Show the plot
plt.show()

# Finding the most frequent movie duration in the 1990s
duration = 95 

# Filtering action movies with duration less than 90 minutes
short_action_movies = movies[
    (movies['duration'] < 90) & (movies['genre'] == 'Action')
]

print(short_action_movies)

#Count the number of short action movies released in the 1990s and save this integer as short_movie_count.
short_movie_count = len(short_action_movies)
print("Count the number of short action movies released in the 1990s and save this integer as short_movie_count.")
print("short_movie_count: " + str(short_movie_count))
```
